```python
#choose the sub folder which contains the mrc files
data_origin_folder = 'F:\\3ded_kenneth_harris_samples\\191123_med_kdmh'
sub_folder = '\\bismuth_susalicylate_B_20191124_175237'

data_origin_folder = data_origin_folder + sub_folder
print(data_origin_folder)

#folder to write the tifs to
destination_folder = "E:\\hl585\\duncan_data_processed\\3ded_kenneth_harris_samples\\191123_med_kdmh\\bismuth_susalicylate_B_20191124_175237"
```

```python
filenames = []
print(data_origin_folder)
for root, dirs, files in os.walk(data_origin_folder):
    for file in files:
        if file.endswith(".mrc"):
             filenames.append(os.path.join(root, file))

filenames = np.sort(filenames)

print(filenames)
```
